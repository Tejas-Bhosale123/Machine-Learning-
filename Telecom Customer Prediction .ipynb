```python
# If df is not a DataFrame, an AssertionError will be raised
# Filter Out Rows with Infinite Values


def clean_dataset(df):
    assert isinstance(df, pd.DataFrame)
    df.dropna(inplace = True)
# Filter Out Rows with Infinite Values
# .any(1) checks if any of these values are present in each row, returning True for rows containing NaN, inf, or -inf
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
# df[indices_to_keep] filters the DataFrame, retaining only rows that do not contain NaN, inf, or -inf values
# .astype(np.float64) converts all values in the DataFrame to float64
    return df[indices_to_keep].astype(np.float64)
```

```python
# replaces missing values in df by filling them with interpolated values

# method='linear': Fills missing values linearly based on the values directly above and below each missing point.
# method='time': If the index is a datetime, it interpolates based on time.
# method='polynomial': Fits a polynomial curve to fill in the values, which can be useful for more complex patterns.
# method='nearest', method='spline', etc., depending on the pattern of missing data.
# Interpoliation is used to Handling Missing Data, Smoothing Data,Improving Data Resolution
df = df.interpolate()
```

```python
# Data model building
# splitting data into training and testing

from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size = 0.2, random_state = 1111)
```

```python
len(xtrain)
```

```python
xtrain[:10]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
```

```python
model_params = {
    
    'random forest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators':[1,5,10]
        }
    },
    'logistic regression': {
        'model': LogisticRegression(solver='saga',max_iter=500),
        'params':{
            'C':[1,5,10]
        }
    },
    'decision tree': {
        'model': DecisionTreeClassifier(),
        'params':{
            'criterion':['gini','entropy']
        }
    },
    'xgboost':{
        'model': XGBClassifier(),
        'params':{
            'base_score':[0.5]
        }
    },
}
```

```python
# ShuffleSplit is a cross-validation method in machine learning that randomly shuffles and splits 
# the dataset multiple times to create different training and testing sets

from sklearn.model_selection import ShuffleSplit
```

```python
# Getting the best_score from the applied models
# performs hyperparameter tuning using GridSearchCV with ShuffleSplit cross-validation to find the 
# best parameters for different models specified in model_params

from sklearn.model_selection import GridSearchCV
scores = []
cv = ShuffleSplit(n_splits = 5, test_size = 0.2, random_state = 0)

for model_name, mp in model_params.items():
    clf = GridSearchCV(mp['model'],mp['params'], cv = cv, return_train_score = False)
    clf.fit(x,y)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
    
df = pd.DataFrame(scores,columns = ['model', 'best_score_','best_params_'])
df
```
